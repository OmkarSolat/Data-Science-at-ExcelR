# **PCA - Principle Component Analysis (clustering)**



```python
#Importing Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing Dataset
univ = pd.read_csv('/content/Universities.csv')
```

```python
univ.head()
```

```python
univ1 = univ.copy()
```

```python
univ1.head()
```

```python
#univ = univ.iloc[:,1:]
univ.drop('Univ',axis=1,inplace=True)
```

```python
univ
```

# Standardization of the Data

```python
#Standardization

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
univ = ss.fit_transform(univ)
```

```python
univ
```

# Applying PCA

```python
#Importing PCA
from sklearn.decomposition import PCA
```

```python
#Building PCA for 2 components
pc2 = PCA(n_components=2)
```

```python
pc2.fit_transform(univ)
```

```python
p_components = pd.DataFrame(pc2.fit_transform(univ), columns=['PC1','PC2'])
p_components
```

```python
sns.scatterplot(x=p_components['PC1'], y=p_components['PC2'])
```

```python
#Building PCA for all components

pc6 = PCA(n_components=6)
pca_values = pc6.fit_transform(univ)
```

```python
#all the pcs from 1 to 6
pca_values
```

```python
#Variance of pcs
var = pca.explained_variance_ratio_
var
```

```python
#cummulative variance
var1 = np.cumsum(var*100)
var1
```

```python
#weights
pca.components_
```

```python
plt.plot(var1)
plt.show()
```

# Data Reduction

```python
pcs = pd.DataFrame(pca_values[:,0:2],columns=["PC1","PC2"])
pcs
```

```python
Final_df = pd.concat([univ1["Univ"],pcs],axis=1)
Final_df
```

```python
sns.scatterplot(data=Final_df,x='PC1',y='PC2',hue='Univ')
```
