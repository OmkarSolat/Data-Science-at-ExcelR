```python
year = np.arange(2000,2025)
print(year)
temp = np.random.randint(20,30,25)        #(start,end,size)
print(temp)
```

# Using matplotlib

```python
#line plot
fig = plt.figure(figsize=(8,5))
plt.plot(year,temp)
plt.xlim(1999,2025)
plt.ylim(15,35)
plt.xlabel('Year',size=15)
plt.ylabel('Average Temperature', size=20)
plt.title('Comparison of Yearly Average Temperature', size=20)
plt.suptitle('A line Plot', size=25, y=1)
```

* List of Named colors:https://matplotlib.org/stable/gallery/color/named_colors.html

* List of markers:https://matplotlib.org/stable/api/markers_api.html

```python
#colors & markers
fig = plt.figure(figsize=(8,5))
plt.plot(year,temp,'r*-')
plt.xlim(1999,2025)
plt.ylim(15,35)
plt.xlabel('Year',color='g',size=15)
plt.ylabel('Average Temperature',color='g', size=20)
plt.title('Comparison of Yearly Average Temperature', size=20)
plt.suptitle('A line Plot',color='r', size=25, y=1)
```

# Using seaborn

```python
#stripplot
sns.stripplot(y=emp['Age'], size=5, jitter=True)
```

```python
sns.stripplot(x=emp['Department'] , y=emp['Age'] , size=5 , jitter=True,
              palette={'Sales':'r','Research & Development':'g','Human Resources':'b'}
             )
```

```python
#scatterplot
sns.scatterplot(x=emp['Age'], y=emp['TotalWorkingYears'],
                hue=emp['Gender'],
                style=emp['Gender'],
                markers={'Male':'^', 'Female':'^'},
                palette={'Male':'darkgoldenrod','Female':'darkcyan'}
               )
```

```python
#barplot
sns.barplot(x=emp['EducationField'], y=emp['MonthlyIncome'], estimator='min')
```

```python
sns.barplot(x=emp['EducationField'], y=emp['MonthlyIncome'], estimator='min', hue=emp['Gender'])
```

```python
#heatmap
emp_heatmap = emp.pivot_table(index='JobRole', columns='BusinessTravel', values='DailyRate')
sns.heatmap(emp_heatmap)
```
