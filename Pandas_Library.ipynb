```python
#male passengers (kids) with one year old
ship[(ship['age']==1) & (ship['sex']=='male')]
```

```python
#Find how many childrens has survived
len(ship[(ship['age']<10) & (ship['survived']==1)])
```

```python
#Find how many females has survived
len(ship[(ship['sex']=='female') & (ship['survived']==1)])
```

```python
#solo travellers i.e. sibsp==0
ship[(ship['parch']==0) & (ship['sibsp']==0)]
```

```python
#female solo travellers
ship[(ship['parch']==0) & (ship['sibsp']==0) & (ship['sex']=='female')]
```

```python
# Grouping ages with average ticket fare
ship.groupby(['age'])['fare'].mean()
```

```python
#Question
#Grouping solo passengers based on embarkation
ship[(ship['sibsp']==0) & (ship['parch']==0)].groupby('embarked').count()
```

# DataFrame and operations

```python
food = pd.DataFrame({'Fruits':['Banana', 'Plum', 'Mango'],
                     'Vegetables':['Potato', 'Onion', 'Peas'],
                     'Grains':['Rice', 'Wheat', 'Daal'],
                     'Snacks':['Samosa', 'Pizza', 'Burger']})
food
```

```python
food.drop('Snacks', axis=1)         #temporary deletion
```

```python
food
```

```python
food.drop('Snacks', axis=1, inplace=True)         #parmanent deletion
```

```python
food
```

```python
food['Fruits'].replace({'Banana':'Apple'}, inplace=True)
food
```
