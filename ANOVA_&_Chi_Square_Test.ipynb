```python
#Importing Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn import datasets
```

NULL Hypothesis : All sepals-petals are equal

Alternate Hypothesis : Atleast few sepals-petals are diffrent from each other.

```python
df = pd.DataFrame(iris.data)
```

 Conclusion :

Here; p-value (4.660592480454751e-159) is less than 0.05 .So, Null Hypothesis is rejected.Thus atleast few sepals-petals are diffrent from each other.

 Null Hypothesis(H0) : The mean values of the three diet plans are equal (mu1 = mu2 = mu3)

 Alternate Hypothesis(Ha) : The mean values of the three diet plans are not the same (mu1 != mu2 != mu3)

```python
from scipy.stats import f_oneway
```

```python
# Create pandas series for all diet plans
at = pd.Series([6,2,3,4,2,3,3,2,7,8,10,4])
gm = pd.Series([4,4,5,7,8,5,3,7,10,4,5,2])
sb = pd.Series([6,7,5,6,8,7,8,9,6,5,5,12])
```

```python
# Conduct the one-way ANOVA
f_oneway(at,gm,sb)
```

Since; p-value is less than alpha = 0.05, the null hypothesis is rejected. There is significant diffrence between the three diet plans.

# Chi-Square Test of Independence

We want to determine whether there is a statistically significant association between smoking and being a professional athlete. Smoking can only be “yes” or “no” and being a professional athlete can only be “yes” or “no”. The two variables of interest are qualitative variables, so we need to use a Chi-square test of independence, and the data have been collected on 28 persons.

```python
# Create a table
data = [['Athlete', 'Smoker'] for i in range(4)] + \
        [['Athlete', 'Non-Smoker'] for i in range(14)] + \
        [['Non-Athlete', 'Smoker'] for i in range(10)] + \
        [['Non-Athlete', 'Non-Smoker'] for i in range(0)]
```

```python
df = pd.DataFrame(data, columns = ['Athlete_or_not', 'smoker_or_non-smoker'])
```

```python
df
```

```python
df_cross = pd.crosstab(df['Athlete_or_not'],df['smoker_or_non-smoker'], margins=True, margins_name="Total")
df_cross
```

```python
from scipy.stats import chi2_contingency
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create a contingency table using numpy
observed = np.array([[14, 4], [0, 10]])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:", expected)
```

Since; p-value is less than alpha = 0.05, the null hypothesis is rejected. there is a statistically significant association between smoking and being a professional athlete.
