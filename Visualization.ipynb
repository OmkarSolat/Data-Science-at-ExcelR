# seaborn

```python
#Importing seaborn library
import seaborn as sns
```

```python
#seaborn has some built-in datasets
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#stripplot
sns.stripplot(y='tip',data=tips, jitter=True)
plt.ylabel('tip ($)')
```

```python
#Grouping with stripplot
sns.stripplot(x='day', y='tip',data=tips, hue='day')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip',data=tips,jitter=False,hue='day')
plt.ylabel('tip($)')
plt.show()
```

```python
#swarm plot
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tips($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex',orient='h')
plt.ylabel('tips($)')
plt.show()
```

```python
#boxplot
sns.boxplot(x='day',y='tip',data=tips,hue='day')
plt.ylabel('tip($)')
```

```python
#violinplot
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tips($)')
plt.show()
```

```python
#Joint plot
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#Density plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
#Paiplot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

# Titanic Data Visualization

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/train.csv')
```

```python
data.head()
```

```python
#Data types
data.dtypes
```

* Proportion of target (survived)

```python
data.groupby('Survived')['PassengerId'].count()
```

* Visual exploration

```python
sns.boxplot(x='Survived',y='Age',data=data,hue='Survived')
# 0 = not survived
# 1 = survived
```

```python
sns.boxplot(x='Survived',y='Age',data=data,hue='Sex')
```

```python
sns.barplot(x='Survived',y='Age',data=data,hue='Survived')
```

```python
sns.barplot(x='Survived',y='Age',data=data,hue='Sex')
```

```python
sns.barplot(x='Pclass',y='Survived',data=data,hue='Pclass')
```

```python
sns.barplot(x='Pclass',y='Survived',data=data,hue='Sex')
```

```python
sns.barplot(x='SibSp',y='Survived',data=data,hue='SibSp')
```

```python
sns.barplot(x='Parch',y='Survived',data=data,hue='Parch')
```

```python
#Distribution plot
survived = data.loc[data['Survived']==1, 'Age'].dropna()
sns.distplot(survived)
plt.title('Survived')
```

```python
not_survived = data.loc[data['Survived']==0, 'Age'].dropna()
sns.distplot(not_survived)
plt.title('Not Survived')
```

```python
data.head()
```

```python
sns.pairplot(data.dropna())
```

```python
#Pclass vs Survive

grid = sns.FacetGrid(data,col='Survived',row='Pclass',aspect=1.5)
grid.map(plt.hist,'Age',alpha=.5,bins=20)
grid.add_legend()
```

```python
#Survival based on station/boarding
sns.barplot(x='Embarked',y='Survived',data=data,hue='Embarked')
```

```python
sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title('Cherbourg')
plt.xticks([0,1])
plt.show()

sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.title('QueensTown')
plt.xticks([0,1])
plt.show()

sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.title('Southampton')
plt.xticks([0,1])
plt.show()
```

```python
figbi, axesbi = plt.subplots(2,4, figsize=(16,10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x='Survived',y='Age',data=data,ax=axesbi[1,1])
sns.boxplot(x='Survived',y='Fare',data=data,ax=axesbi[1,2])
```
