# **Forecasting : Sampling & Transformations**

# Up-sampling Data

```python
# upsample to daily intervals
import warnings
warnings.filterwarnings("ignore")
from pandas import read_csv
import datetime
from matplotlib import pyplot
```

```python
series = read_csv('sales.csv', header=0, index_col=0, parse_dates=True)
```

```python
series
```

```python
upsampled = series.resample('D').mean()
print(upsampled.head(50))
```

# interpolate the missing value

```python
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(30))
interpolated.plot()
pyplot.show()
```

```python
series.plot()
```

# Down-sampling Data

```python
# downsample to quarterly intervals
from pandas import read_csv
import datetime
from matplotlib import pyplot
```

```python
resample = series.resample('Q')
quarterly_mean_sales = resample.mean()
```

```python
print(quarterly_mean_sales.head())
quarterly_mean_sales.plot()
pyplot.show()
```

```python
quarterly_mean_sales.plot()
```

# Tranformations

```python
# load and plot a time series
from pandas import read_csv
from matplotlib import pyplot
```

```python
#series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/airline-passengers.csv', header=0, index_col=0, parse_dates=True)
series = read_csv("airline-passengers.csv")
series
```

# Square Root Transform

```python
from pandas import read_csv
from pandas import DataFrame
from matplotlib import pyplot
import pandas as pd
import numpy as np
```

```python
dataframe = series
dataframe.head()
```

```python
#dataframe = pd.DataFrame(series.values, columns=["Month", "Passengers"])
#L1 = dataframe["Passengers"].to_list()
#dataframe['passengers'] = np.sqrt(L1)
```

```python
from numpy import sqrt
dataframe = pd.DataFrame(series.values)
dataframe.columns = ['Month','passengers']
L1 = dataframe['passengers'].tolist()
dataframe['passengers']= np.sqrt(L1)
```

```python
dataframe
```

```python
# line plot
pyplot.subplot(311)
pyplot.plot(dataframe['passengers'])

pyplot.subplot(312)
pyplot.plot(dataframe['passengers'])

# histogram
pyplot.subplot(313)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```

# Log Transform

```python
from numpy import log
dataframe = pd.DataFrame(series.values)
dataframe.columns = ['Month','passengers']
L2 = dataframe['passengers'].tolist()
dataframe['passengers']= np.log(L2)
```

```python
dataframe
```

```python
# line plot
pyplot.subplot(211)
pyplot.plot(dataframe['passengers'])

# histogram
pyplot.subplot(212)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```
